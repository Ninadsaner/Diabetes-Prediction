```python
def vif_score(x):
    scalar = StandardScaler()
    arr = scalar.fit_transform(x)
    return pd.DataFrame([[x.columns[i], variance_inflation_factor(arr,i)] for i in range(arr.shape[1])], columns=['FEATURE', 'SCORE'])
```

```python
vif_score(X)
```

### VIF 1 = No corellation
### 1<VIF<5 = Moderate correlation
### VIF> = High correlation
